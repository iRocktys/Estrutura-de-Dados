## Contagem de Pares

A execução desse exercício é baseado no código de recursão do MaxMin, o objetivo é olhar o funcionamento da divisão e conquista

```python
%%cython
from libc.stdlib cimport rand, srand
from libc.time cimport time

cdef int contar_pares(int[:] A, int linf, int lsup):
    if linf == lsup:
        return 1 if A[linf] % 2 == 0 else 0

    cdef int meio = (linf + lsup) // 2
    cdef int count1 = contar_pares(A, linf, meio)
    cdef int count2 = contar_pares(A, meio + 1, lsup)
    return count1 + count2

# Criar vetor com 15 elementos aleatórios
srand(<unsigned int> time(NULL))
cdef int i
cdef int[15] vetor

for i in range(15):
    vetor[i] = rand() % 20

print("Vetor:", [vetor[j] for j in range(15)])
print("Pares:", contar_pares(vetor, 0, 14))
```

### Contagem de Impares e pares

```python
%%cython
from libc.stdlib cimport rand, srand
from libc.time cimport time

cdef tuple contar_pares_impares(int[:] A, int linf, int lsup):
    if linf == lsup:
        if A[linf] % 2 == 0:
            return (1, 0)
        else:
            return (0, 1)

    cdef int meio = (linf + lsup) // 2
    cdef tuple c1 = contar_pares_impares(A, linf, meio)
    cdef tuple c2 = contar_pares_impares(A, meio + 1, lsup)
    return (c1[0] + c2[0], c1[1] + c2[1])

# Criar vetor com 15 elementos aleatórios
srand(<unsigned int> time(NULL))
cdef int i
cdef int[15] vetor

for i in range(15):
    vetor[i] = rand() % 20

pares, impares = contar_pares_impares(vetor, 0, 14)
print("Vetor:", [vetor[j] for j in range(15)])
print("Pares:", pares)
print("Ímpares:", impares)
```
