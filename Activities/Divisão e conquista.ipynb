# Divisão e conquista

**Enunciado.** Dado um vetor `A` com `n` números inteiros, projetar um algoritmo baseado em divisão e conquista para determinar quantos elementos pares existem no vetor.

O notebook apresenta pseudocódigo conceitual, implementação em Cython, recorrência e análise assintótica.

```python
%load_ext cython
```

## Ideia do algoritmo

A estratégia segue os três passos clássicos:

1. **Dividir:** separar o vetor em duas metades;
2. **Conquistar:** resolver recursivamente cada metade;
3. **Combinar:** somar a quantidade de pares obtida nas duas metades.

O caso base ocorre quando o intervalo tem um único elemento. Nesse caso, basta verificar se ele é par.

## Recorrência

A cada chamada, o problema é dividido em dois subproblemas de tamanho aproximadamente `n/2`. A combinação é apenas uma soma, ou seja, custo constante:

\[
T(n) = 2T(n/2) + \Theta(1)
\]

Como há `n` folhas na árvore de recursão e cada elemento é verificado uma única vez, a complexidade total é:

\[
T(n) = \Theta(n)
\]

O espaço de pilha é **Θ(log n)** para divisões balanceadas.

```python
%%cython
# cython: boundscheck=False, wraparound=False, cdivision=True

cdef long _contar_pares_dc(list A, Py_ssize_t esquerda, Py_ssize_t direita):
    cdef Py_ssize_t meio
    cdef long esq, dir

    if esquerda > direita:
        return 0

    if esquerda == direita:
        if (<long>A[esquerda]) % 2 == 0:
            return 1
        return 0

    meio = (esquerda + direita) // 2
    esq = _contar_pares_dc(A, esquerda, meio)
    dir = _contar_pares_dc(A, meio + 1, direita)
    return esq + dir

def contar_pares_divisao_conquista(list A):
    return _contar_pares_dc(A, 0, len(A) - 1)
```

```python
A = [10, 3, 5, 8, 12, 7, 4]
print(A)
print("Quantidade de pares:", contar_pares_divisao_conquista(A))
```

## Teste de mesa resumido

Para `A = [10, 3, 5, 8, 12, 7, 4]`:

- metade esquerda: `[10, 3, 5, 8]` possui `2` pares;
- metade direita: `[12, 7, 4]` possui `2` pares;
- combinação final: `2 + 2 = 4`.

O algoritmo não ordena o vetor e não altera os dados. Ele apenas divide os intervalos e soma contagens.

```python
assert contar_pares_divisao_conquista([1, 3, 5]) == 0
assert contar_pares_divisao_conquista([2, 4, 6]) == 3
assert contar_pares_divisao_conquista([]) == 0
print("Testes concluídos com sucesso.")
```
