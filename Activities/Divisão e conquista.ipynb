# Divisão e conquista

## MaxMin

```python
import pandas as pd

class ContadorMaxMin:
    def __init__(self):
        self.chamadas = 0
        self.comparacoes = 0
        self.passos = []


def max_min(A, inf=0, sup=None, contador=None):
    if sup is None:
        sup = len(A) - 1
    if contador is None:
        contador = ContadorMaxMin()

    contador.chamadas += 1

    if inf == sup:
        contador.passos.append({'chamada': f'MAX_MIN({inf}, {sup})', 'subvetor': [A[inf]], 'max': A[inf], 'min': A[inf]})
        return A[inf], A[inf], contador

    if sup - inf == 1:
        contador.comparacoes += 1
        if A[inf] < A[sup]:
            minimo, maximo = A[inf], A[sup]
        else:
            minimo, maximo = A[sup], A[inf]
        contador.passos.append({'chamada': f'MAX_MIN({inf}, {sup})', 'subvetor': A[inf:sup+1], 'max': maximo, 'min': minimo})
        return maximo, minimo, contador

    meio = (inf + sup) // 2
    max1, min1, contador = max_min(A, inf, meio, contador)
    max2, min2, contador = max_min(A, meio + 1, sup, contador)

    contador.comparacoes += 2
    maximo = max1 if max1 > max2 else max2
    minimo = min1 if min1 < min2 else min2

    contador.passos.append({'chamada': f'MAX_MIN({inf}, {sup})', 'subvetor': A[inf:sup+1], 'max': maximo, 'min': minimo})
    return maximo, minimo, contador
```

## Execução MaxMin

```python
A = [9, 3, 10, 12, 5]
maximo, minimo, contador = max_min(A)

print('Vetor:', A)
print('Máximo:', maximo)
print('Mínimo:', minimo)
print('Chamadas:', contador.chamadas)
print('Comparações:', contador.comparacoes)

display(pd.DataFrame(contador.passos))
```

## Contagem de pares e ímpares

```python
class ContadorParImpar:
    def __init__(self):
        self.chamadas = 0
        self.passos = []


def conta_pares_impares(A, linf=0, lsup=None, contador=None):
    if lsup is None:
        lsup = len(A) - 1
    if contador is None:
        contador = ContadorParImpar()

    contador.chamadas += 1

    if linf == lsup:
        pares = 1 if A[linf] % 2 == 0 else 0
        impares = 1 - pares
        contador.passos.append({'chamada': f'CONTA({linf}, {lsup})', 'subvetor': [A[linf]], 'pares': pares, 'ímpares': impares})
        return pares, impares, contador

    meio = (linf + lsup) // 2
    pares1, impares1, contador = conta_pares_impares(A, linf, meio, contador)
    pares2, impares2, contador = conta_pares_impares(A, meio + 1, lsup, contador)

    pares = pares1 + pares2
    impares = impares1 + impares2
    contador.passos.append({'chamada': f'CONTA({linf}, {lsup})', 'subvetor': A[linf:lsup+1], 'pares': pares, 'ímpares': impares})
    return pares, impares, contador
```

## Execução pares e ímpares

```python
A = [4, 7, 2, 9, 10]
pares, impares, contador = conta_pares_impares(A)

print('Vetor:', A)
print('Pares:', pares)
print('Ímpares:', impares)
print('Chamadas:', contador.chamadas)

display(pd.DataFrame(contador.passos))
```
