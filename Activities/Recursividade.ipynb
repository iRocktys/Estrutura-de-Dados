# Recursividade

**Problema.** Verificar recursivamente se um vetor é simétrico, isto é, se ele lê da mesma forma da esquerda para a direita e da direita para a esquerda.

Exemplos:

```python
[4, 7, 9, 7, 4]  -> verdadeiro
[3, 5, 5, 2]     -> falso
```

A solução não usa laços iterativos.

````python
%load_ext cython
````

## Ideia recursiva

A função recebe:

- o vetor `A`;
- um índice esquerdo `i`;
- um índice direito `j`.

A cada chamada:

1. se `i >= j`, o intervalo terminou ou chegou ao centro: retorna verdadeiro;
2. se `A[i] != A[j]`, a simetria é quebrada: retorna falso;
3. caso contrário, chama a função para o intervalo interno `i + 1, j - 1`.

## Análise assintótica

Cada chamada compara no máximo um par de elementos. O intervalo diminui em duas posições a cada chamada.

A recorrência é:

\[
T(n) = T(n-2) + \Theta(1)
\]

Logo, o tempo é **Θ(n)**.

A profundidade da recursão é aproximadamente `n/2`, então o espaço de pilha é **Θ(n)**.

````python
%%cython
# cython: boundscheck=False, wraparound=False, cdivision=True

cdef bint _simetrico_rec(list A, Py_ssize_t i, Py_ssize_t j):
    if i >= j:
        return True
    if A[i] != A[j]:
        return False
    return _simetrico_rec(A, i + 1, j - 1)

def eh_simetrico(list A):
    return bool(_simetrico_rec(A, 0, len(A) - 1))
````

````python
testes = [
    [4, 7, 9, 7, 4],
    [3, 5, 5, 2],
    [1, 2, 2, 1],
    [1],
    [],
]

for t in testes:
    print(t, "=>", eh_simetrico(t))
````

## Perguntas da atividade

**Qual é o caso base da recursão?**  
Quando `i >= j`, pois o algoritmo já comparou todos os pares necessários.

**Em que situação a função encerra imediatamente com falso?**  
Quando `A[i] != A[j]`.

**Como os parâmetros mudam a cada chamada?**  
O índice esquerdo aumenta e o direito diminui: `i + 1` e `j - 1`.
