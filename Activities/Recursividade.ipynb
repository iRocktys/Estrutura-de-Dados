# Recursividade

## Vetor simétrico

```python
import pandas as pd

class ContadorSimetrico:
    def __init__(self):
        self.chamadas = 0
        self.passos = []


def simetrico(A, i=0, j=None, contador=None):
    if j is None:
        j = len(A) - 1
    if contador is None:
        contador = ContadorSimetrico()

    contador.chamadas += 1

    if i >= j:
        contador.passos.append({'chamada': f'SIMETRICO({i}, {j})', 'comparação': '-', 'resultado': True})
        return True, contador

    igual = A[i] == A[j]
    contador.passos.append({'chamada': f'SIMETRICO({i}, {j})', 'comparação': f'{A[i]} == {A[j]}', 'resultado': igual})

    if not igual:
        return False, contador

    return simetrico(A, i + 1, j - 1, contador)
```

## Execução

```python
exemplos = [
    [4, 7, 9, 7, 4],
    [3, 5, 5, 2],
    [1, 2, 2, 1],
    [8]
]

linhas = []
for A in exemplos:
    resultado, contador = simetrico(A)
    linhas.append({'A': A, 'Simétrico': resultado, 'Chamadas': contador.chamadas})

display(pd.DataFrame(linhas))
```

## Passo a passo

```python
A = [4, 7, 9, 7, 4]
resultado, contador = simetrico(A)

print('Vetor:', A)
print('Resultado:', resultado)
print('Chamadas:', contador.chamadas)

display(pd.DataFrame(contador.passos))
```
